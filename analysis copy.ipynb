```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configure visualizations for inline display (if using Jupyter Notebook)
%matplotlib inline

# --- Step 1: Load the Data ---
# Replace 'data.csv' with your actual file name/path.
# df = pd.read_csv('data.csv')

# Quick look at the dataset
print("Dataset Info:")
print(df.info())
print("\nFirst 5 rows:")
print(df.head())

```

```python
# --- Step 2: Data Cleaning and Pre-processing ---

# Convert columns to appropriate numeric types.
# Specify columns to convert (excluding categorical/text columns).
numeric_cols = [
    'year', 'gov_exp_pct_gdp', 'lit_rate_adult_pct', 'pri_comp_rate_pct', 
    'pupil_teacher_primary', 'pupil_teacher_secondary', 
    'school_enrol_primary_pct', 'school_enrol_secondary_pct', 
    'school_enrol_tertiary_pct'
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Check how many missing values exist in each column
print("\nMissing Values per Column:")
print(df.isnull().sum())

# Option 1: Impute missing numeric values with the median 
# (this is one common strategy; other approaches include dropping or interpolating)
for col in numeric_cols:
    median_value = df[col].median()
    df[col].fillna(median_value, inplace=True)

```

```python
country_name = 'Afghanistan'
country_df = df[df['country'] == country_name]

plt.figure(figsize=(10, 6))
plt.plot(country_df['year'], country_df['gov_exp_pct_gdp'], marker='o', linestyle='-')
plt.title(f'Government Expenditure (% GDP) Over Time - {country_name}')
plt.xlabel('Year')
plt.ylabel('Gov Exp (% GDP)')
plt.grid(True)
plt.show()

# Visualization 2: Scatter Plot - Adult Literacy Rate vs Primary Completion Rate
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='lit_rate_adult_pct', y='pri_comp_rate_pct', 
                hue='year', palette='viridis', alpha=0.7)
plt.title('Adult Literacy Rate vs Primary Completion Rate')
plt.xlabel('Adult Literacy Rate (%)')
plt.ylabel('Primary Completion Rate (%)')
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Visualization 3: Correlation Heatmap among Numeric Variables
plt.figure(figsize=(12, 10))
corr_matrix = df[numeric_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numeric Variables')
plt.show()

# Visualization 4: Distributions of Pupil-Teacher Ratios (Primary & Secondary)
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df['pupil_teacher_primary'].dropna(), bins=30, kde=True)
plt.title('Distribution of Primary Pupil-Teacher Ratio')
plt.xlabel('Pupil-Teacher Ratio (Primary)')

plt.subplot(1, 2, 2)
sns.histplot(df['pupil_teacher_secondary'].dropna(), bins=30, kde=True)
plt.title('Distribution of Secondary Pupil-Teacher Ratio')
plt.xlabel('Pupil-Teacher Ratio (Secondary)')

plt.tight_layout()
plt.show()
```
